```python
# Integers: Addition Example
# Addition
sum = 3 + 7

# Result: print the sum
# To show a variable, we use the print() function
print(sum)
```

```python
# Integers: Subtraction Example
# Subtraction
min = 10 - 4

# Result: print the subtraction
# To show a variable, we use the print() function
print(min)
```

```python
# Integers: Multiplication Example
# Multiplication
mul = 4 * 6

# Result: print the multiplication
# To show a variable, we use the print() function
print(mul)
```

```python
# Integers: Division Example
# Division
div = 20 / 5

# Result: print the division
# To show a variable, we use the print() function
print(div)
```

```python
## Strings
# If we want to store text, we use strings.
# We can create strings by inserting an ''='' and consequently the text is going to b in the middle of ' '
name1 = 'Pasquale'
# We can use also double quotes " "
secondName = "Riccardo"

name1 = "Pasquale"

# To show a string, we use the print() function
print(secondName)
print(name1)
```

```python
# We can include strings and Integers in a single print statement

names = "Pasquale, Renato, Veronica"
var1 = 24
var2 = 20
var3 = 15

print("We are" ,names, "and we are", var1, var2, var3, "years old.")
```

```python
# We can also check the type of item we're typing, by using the type() function.
name5 = 'Hello'
age = 15

print(type(name5))
print(type(age))
```

```python
# Lists and Indexables Integers and Strings
vegetables = ["carrot", "broccoli", "spinach", "cauliflower", "bellpepper", "lettuce", "cabbage"]
# If we want to print a specific item from the list, we can index them, by using [] and inserting the number, as the position the item has in the list.
print(vegetables[0])
```
