```python
# Augmented Assignment Operators
# We can use augmented assignment operators to modify the value of a variable in a more concise way.
# For example, instead of writing ab = ab + 3, we can use the += operator to achieve the same result.
ab = 5
ab += 3 
ab += 4
print (ab)

# Another example with subtraction
bc = 10
bc -= 2
bc -= 3
print(bc)

# Another example with multiplication
cd = 4
cd *= 3
cd *= 2
print(cd)

# Another example with division
de = 20
de /= 4
de /= 2
print(de)
```

```python
# Lists
# Lists are a collection of items that can be of different types, such as integers, strings, or even other lists.
# We can create a list by using square brackets [] and separating the items with commas.
my_list = [1, 2, 3, "hello", "world"] #As you can see, the list can contain both inegers, strings and even other types of items.

# We can access the items in a list by using their index, which starts at 0.
print(my_list[0]) # This will print the first item in the list, which is, in this case, the integer 1.

# We can also use negative numbers to index lists.
print(my_list[-1]) # This will print the last item in the list, which is, in this case, the string "world".

# We can also create some "nested" lists, so you can assign some items to other items in the list.
# For instance:
bed = 1.56
kit = 2.34
bath = 6.42
gard = 15.13 # we want to assign to each of these listed variables, a name.


house = [
    ["bedroom", bed],
    ["kitchen", kit],
    ["bathroom", bath],
    ["garden", gard]
]

print(house) # In this case, it will print a list with 4 items, each of them have a name and an integer assigned to it.
```

```python
# Update the value of an item to the list.
areas = [
    "hallway", 11.25,
    "kitchen", 18.0,
    "living room", 20.0,
    "bedroom", 10.75,
    "bathroom", 9.50
]

# For this list, we want to change the value of the bathroom area.
areas[-1] = 10.50 # We can do this by indexing the item we want to change and assigning it a new value.
print(areas) # by printing the list, it will show the updated version.
# This because the list is being used as a ''type'' of value, and whenever we change the valuse that are in the list, it will update them automatically..
```

```python
# Add another item to the list.
# There are three ways to add an item:
# 1. Using the append() function
areas.append("pool")
print(areas) # by printing the list, it will show the updated version.
# In this case, you will be able to add only one item at a time, but if you want to add more than one item.

# 2. Using the extend() function
areas.extend(["garage", 12.0]) # This will add both the name and the value of the garage to the list.
print(areas) # by printing the list, it will show the updated version.

# 3. Using the mathematical operator +
areas_1 = areas + ["garden", 26.45]
print(areas_1) # by printing the list, it will show the updated version.
```

```python
# Remove an item from the list.
# There's one way to remove an item:

# Using the function del
del areas[8] # This will remove the item "garage" from the list, which is in the position 8.
del areas[8] # This will remove the item 12.0 from the list, which is in the position 8, because after removing "garage", the item 12.0 is now in the position 8.
print(areas) # by printing the list, it will show the updated version.
```
